## IRIS DATA FRAME

```python
 import plotly.express as px
```

```python
df=px.data.iris()
df.head()
```

```python
df['species'].unique()
```

## SCATTER PLOT

```python
fig=px.scatter(df,x='sepal_length',y='sepal_width',color='species',marginal_x='violin',marginal_y='box')
fig.show()
```

## REGRESSION PLOT

```python
fig=px.scatter(df,x='sepal_length',y='sepal_width',color='species',trendline='ols') #ordinary least square(ols)
fig.show()
```

```python
fig=px.scatter(df,x='sepal_length',y='sepal_width',color='species',trendline='ols',template='simple_white') #ordinary least square(ols) #template changes the background
fig.show()
```

```python
fig1=px.bar(df,x='sepal_length',y='petal_length',color='species')
fig1.show()
```

```python
df1=sns.load_dataset('tips')
df1.head()
```

```python
fig2=px.bar(df1,x='sex',y='total_bill',color='smoker')  #horizontal division
fig2.show()
```

```python
fig2=px.bar(df1,x='sex',y='total_bill',color='smoker',barmode='group')  #vertical division
fig2.show()
```

```python
fig2=px.bar(df1,x='sex',y='total_bill',color='smoker',barmode='group',pattern_shape='smoker',pattern_shape_sequence=['x','.'])  #vertical division
fig2.show()
```

## SCATTER MATRIX

```python

  
fig=px.scatter_matrix(df,dimensions=['sepal_length','sepal_width','petal_length','petal_width'],color='species',width=800,height=500)
fig.show()
```

## LINE PLOT

```python
fig=px.line(df,x='sepal_length',y='sepal_width',color='species',width=800,height=500)  # forms straight line
fig.show()
```

```python
fig=px.line(df,x='sepal_length',y='sepal_width',color='species',line_shape='spline')  # forms curved lines
fig.show() 
```

```python
fig=px.pie(df,names='species',values='petal_length')
fig.show()
```

```python
fig=px.pie(df1,names='day',values='tip')
fig.show()
```

## HISTOGRAM

```python
fig=px.histogram(df,x='petal_length',y='petal_width',color='species',title='petal_length vs petal_width')
fig.show()
```

```python
fig=px.histogram(df1,x='total_bill',y='tip',color='sex',title='total_bill vs tips')
fig.show()
```

## VIOLIN PLOT

```python
fig=px.violin(df,x='petal_length',y='species',color='species')
fig.show()
```

```python
fig=px.violin(df1,x='day',y='total_bill',color='smoker')
fig.show()
```

```python
fig=px.violin(df1,x='day',y='total_bill',color='smoker',box=True,width=1200,height=800,points='all')
fig.show()
```

## BOX PLOT

```python
fig=px.box(df,x='petal_length',y='species',color='species')
fig.show()
```

```python
fig=px.box(df1,x='day',y='total_bill',color='smoker')
fig.show()
```

```python
fig=px.box(df1,x='day',y='total_bill',color='smoker',notched=True)
fig.show()
```

## DENSITY PLOT

```python
fig=px.density_contour(df,x='sepal_length',y='sepal_width',color='species')
fig.show()
```

## 3-D SCATTER PLOT

```python
fig= px.scatter_3d(df,x='sepal_length',y='sepal_width',z='petal_length',title='3D scatter plot',color='species',height=800,width=600)
fig.show()
```

```python
fig= px.scatter_3d(df1,x='total_bill',y='tip',z='size',title='3D scatter plot',color='sex',height=800,width=600)
fig.show()
```
