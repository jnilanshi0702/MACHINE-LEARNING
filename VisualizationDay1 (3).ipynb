```python
plt.figure(figsize=(7,5)) # change the distance between two points on both axis
a=np.array([1,2,3,4,5,6,7])
b=np.power(a,3)
plt.plot(a,b)  # used to plot a and b
plt.title('my first graph') # used to give title to the graph
plt.xlabel('it is x axis') #used to give name to x axis
plt.ylabel('it is y axis') # used to give name to y axis
plt.show()      # used to show the graph
```

```python
x=np.linspace(0,10,1000)

y=np.sin(x)
z=np.cos(x)
plt.plot(x,y) #used to plot x wrt to y
plt.plot(x,z) # used to plot x wzrt to z
plt.title('sin graph/cosine graph') # used to give title to the graph
plt.xlabel(' x axis') #used to give name to x axis
plt.ylabel(' y axis') # used to give name to y axis
plt.show()
```
