# PIVOT TABLE

```python
data={'Country':['india','us','us','russia','china','us','india'],
      'Year':['2010','2015','2010','2015','2015','2020','2015'],
      'GDP/LITR':['gdp','gdp','gdp','litr','litr','litr','gdp'],
      'Values':[52,56,54,53,59,57,51]}
```

```python
n3=pd.DataFrame(data)
```

```python
n3
```

```python
pd.pivot_table(n3,index=['Country'], aggfunc='sum')
```

```python
pd.pivot_table(n3,index=['Country','Year'], aggfunc='sum')
```

```python
pd.pivot_table(n3,index=['Country'], aggfunc='mean')
```

```python
pd.pivot_table(n3,index=['Country'], aggfunc='max') #gives the max litr/gdp of that country
```

```python
pd.pivot_table(n3,index=['Country'], aggfunc='min') #gives the min litr/gdp of that country
```
