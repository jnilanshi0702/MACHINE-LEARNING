# SEABORN

* Categorical plot----needs atleast one categorical(ex-sex,smoker column in tips dataset) column
1. countplot
2. boxplot
3. barplot
4. swarmplot
5. violinplot
6. 

* Distribution Plot---needs atleast one continuos numerical column(eg. tips,total_bill)
1. distplot
2. histogram
3. kde plot
4. line plot
5. regplot
6. jointplot

* Matrix---needs continuos numerical or categorical(numeric)(eg-size column in tips dataset-it has categorical value in numeric form) column
1. pairplot
2. heatmap plot(important)

* Regression Plot---two continuous column
1. scatter plot
2. regplot

## 2. Heatmap (important)

```python
_=sns.heatmap(df1.corr(),annot=True) 
# correlation tells the relation between two columns,there similarities,how they are related to each other,its value range from -1 to +1,
#if corr=1[POSITIVE CORRELATION] this means they are highlly correlated(or they are similar),with the increase in x, y increases propotionally
#if corr=-1[NEGATIVE CORRELATION] (they are negatively correlated), with the increase in x, y decreases propotionally
#if corr=0.976[NO CORRELATION] (highly nearly correlated) x and y has no relation scatterred data points
# annot=True is used to see the correlation value in the boxes
```

```python
df1.corr()  # gives the correlation value between various columns of the dataset
```

```python
_=sns.heatmap(df1.corr())  #if annot=True is not written no value is given in the corellation boxes
```
