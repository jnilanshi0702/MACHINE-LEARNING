# CONCAT
* USED WHEN WE HAVE SAME COLUMNS DIFFERENT VALUE

```python
data={'id':['1','2','3','4','5','6','7','8'],
      'name':['rachit','mansi','riddhi','isha','ishan','rehan','ruhan','rashi'],
      'age':['31','12','34','16','13','15','17','19',]
          }
```

```python
d1=pd.DataFrame(data) #DATAFRAME 1
```

```python
d1
```

```python
data1={'id':['9','10','11','12','13','14','15','16'],
      'name':['rachit','mansi','riddhi','isha','ishan','rehan','ruhan','rashi'],
      'age':['31','12','34','16','13','15','17','19',]
          }
```

```python
d2=pd.DataFrame(data1) #DATAFRAME 2
```

```python
d2
```

```python
data2={'id':['17','18','19','20','21','22','23','24'],
      'name':['rachit','mansi','riddhi','isha','ishan','rehan','ruhan','rashi'],
      'age':['31','12','34','16','13','15','17','19',]
          }
```

```python
d3=pd.DataFrame(data2) #DATAFRAME 3
```

```python
d3
```

```python
# CONCAT
final_d=pd.concat([d1,d2,d3]) # this command is used to concat the above 3 dataframe
```

```python
final_d  #here the problem is that the index repeats itself for each dataframe
```

```python
final_d.set_index('id') # this will solve the above problem as now the id will become the index so no repeatation
```

```python
final_d=pd.concat([d1,d2,d3],axis=1) # this command shows the same index but the column of each dataframes are written seperately
```

```python
final_d
```

```python
final_d1=pd.concat([d1,d2,d3],ignore_index=True) # this command arranges the index properly no repeatation
```

```python
final_d1
```

# MERGE
* USED WHEN WE HAVE SAME COLUMNS SAME VALUE

```python
data3={'id':['1','2','3','4','5','6','7','8'],
      'name':['rachit','mansi','riddhi','isha','ishan','rehan','ruhan','rashi'],
      'age':['31','12','34','16','13','15','17','19',]
          }
```

```python
dd1=pd.DataFrame(data3) #new dataframe 1
```

```python
data4={'id':['1','2','3','4','5','6','7','8'],
      'name':['rachit','mansi','riddhi','isha','ishan','rehan','ruhan','rashi'],
      'phone':['947454523','545422245','8454231212','543422123','5433666','543646769','57643543545','545565365']
          }
```

```python
dd2=pd.DataFrame(data4) #new dataframe 2
```

```python
a=pd.merge(dd1,dd2,on=['id','name']) #this command merge the column of both the dataframe according to id and name
```

```python
a
```

```python
b=pd.merge(dd1,dd2,on=['id'])#this command merge the column of both the dataframe according to id 
```

```python
b
```

# Group by

```python
data7={'id':['1','2','3','4','5','6','7','8'],
      'company':['google','microsoft','fb','fb','google','microsoft','fb','google'],
      'sales':[250000,340000,368000,34677,38485,67453,12300,340008]
          }
```

```python
dd5=pd.DataFrame(data7)
```

```python
dd5
```

```python
comp_wise=dd5.groupby('company') #this command groups the values company wise
```

```python
comp_wise
```

```python
comp_wise.sum() #this command retrieve the sum of sales company wise
```

```python
comp_wise.max() #this command retrieve the max of sales of each company
```

```python
comp_wise.min() #this command retrieve the min of sales of each company
```

```python
dd5.groupby('company').describe() # this command describes the dataframe company wise
```

```python
comp_wise.describe()   # this command describes the dataframe company wise
```
